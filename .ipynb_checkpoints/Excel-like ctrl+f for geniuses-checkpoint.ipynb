# Tabular Data
## Processing tabular data using numpy

```python
import numpy
import os
```

```python
help(numpy.genfromtxt)
#Delimiter argument: 
#Skip headers: skips lines
```

```python
dist_file= os.path.join('data', 'distance_data_headers.csv')
#genfrom txt: loads data from text file
# fname=filename
# delimiter is splitting values at each comma 
# dtype= reads everything as words
# Outputs a numpy array
distances= numpy.genfromtxt(fname=dist_file, delimiter=',', dtype='unicode')
print(distances)
```

```python
# Headers
headers= distances[0]
print(headers)
```

```python
#Numerical Values
data=distances[1:]
print(data)
```

```python
#change data type of Numerical Values to 'actual numbers'
data= data.astype(numpy.float)
print(data)
```

```python
# Slicing up a 2D list
# array_name[row_index, column_index]

print(data[0,1])
print(data[1,0])
```

```python
#Cut out pieces of data
small_data= data[0:10, 0:3]
print(small_data)
```

```python
# Can cut up further integrated in the print command
print(small_data[:, 1:])
```

```python
# Taking the mean of a dataset 
# data_average= numpy.mean(data_set)
```

```python
print(headers)
```

```python
# One column average
thr4_atp=data[:,1]
avg_thr4_atp= numpy.mean(thr4_atp)
print(avg_thr4_atp)
```

```python
# multiple averages
# range function
num_column=len(headers)
print(num_column)
```

```python
#range fxn
for i in range(1, num_column):
    #print(i)
    column=data[:,i]
    avg_col=numpy.mean(column)
    data_label = headers[i]
    print(F'{data_label} : {avg_col}')
```

# Plotting Tabular Data
## Plotting Data Using matplotlib

```python
#Importing matplotlib
import matplotlib.pyplot as plt
```

```python
# Make new figure
plt.figure()
#Adding data to figure
plt.plot(data[:,1])
```

```python
#Labeling a plot
plt.figure()

sample_name = headers[1]
plt.xlabel('Simulation Frame')
plt.ylabel('Distance(angstrom)')
plt.title("Fake RMSD")
plt.plot(data[:,1], label= sample_name)
plt.legend()

plt.savefig(F'{sample_name}.png', dpi=300)
```

```python
#Do it again, but with two sets of data
plt.figure()

sample_name = headers[1]
plt.xlabel('Simulation Frame')
plt.ylabel('Distance(angstrom)')
plt.title("Fake RMSD")
plt.plot(data[:,1], label= sample_name, color='red')
plt.plot(data[:,2], label=headers[2], color='blue')
plt.legend()

plt.savefig(F'2samples.png', dpi=300)
```

```python
#Making an ugly gross for loop for this

plt.figure()
for col in range(1, num_column):
    plt.plot(data[:, col], label=headers[col])
    plt.xlabel('Simulation Frame')
    plt.ylabel('Distance (angstrom)')
    plt.legend()
    
    plt.savefig('whatever.png')
```

```python
# Make like five hundred diff plots


data_min= numpy.min()

for col in range(1, num_column):
    plt.figure()
    plt.plot(data[:, col], label=headers[col], color='blue')
    plt.ylim(4, 20)
    plt.xlabel('Simulation Frame')
    plt.ylabel('Distance (angstrom)')
    plt.legend()
    plt.savefig(F'{headers[col]}.png')
```

```python
#Finding mins and maxes of data
data_min= numpy.min(data[:,1:])
data_max= numpy.max(data[:,1:])

print(data_min, data_max)
```
